```python
import pandas as pd
import numpy as np

Xmatch = pd.read_csv('/Users/87steven/Documents/ASIAA/Blazar SED code and data/data/BZCAT_Xmatch_ACC_csv.csv') # # = 37671
FIRST = pd.read_csv('/Users/87steven/Documents/ASIAA/Blazar SED code and data/data/FIRST_csv.csv')  # # = 8499

##please add your destination path to save .csv file
SAVEpath = '/Users/87steven/Documents/ASIAA/Blazar SED code and data/FIRST match'   

def ClearData():
    data = {'Name': [],
    'RA': [],
    'DEC': [],
    'z': [], 
    'Rmag': [],
    'class': [],
    'flux_ratio': [],
    'flux_143': [], 
    'flux_xray': [],
    'flux_gamma': [],
    'Spec_index': [], 
    'flux_ALMA': [], 
    'errflux': [], 
    'band': [],
    'freq': [],
    'date': [],
    'Fpeak': [],
    'Fint': [],
    'rms_error': [] }
    return data
```
